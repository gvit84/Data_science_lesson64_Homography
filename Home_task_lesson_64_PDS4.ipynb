```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_src = cv2.imread('chess_1.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
pts_src = np.array([[210, 79], [447, 72], [443, 230], [214, 219]])
```

```python
im_dst = cv2.imread('chess_2.png')
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show()
```

```python
pts_dst = np.array([[222, 312], [217, 67], [471, 115], [382, 347]])
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
h
```

```python
im_out = cv2.warpPerspective(im_src, h,
                            (im_dst.shape[1],
                            im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,6))
cv2.imwrite('result.jpg', im_out)
plt.imshow(im_out)
plt.show()
```
